```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data = {
#     "West Java":    [0, 11, 5, 10, 7, 48, 88, 38, 32],
#     "Central Java": [2,  0, 2, 10, 6, 29, 61, 30, 23],
#     "East Java":    [4,  7, 0,  7, 5, 30, 49, 24, 15],
#     "Jkt_1":        [0,  4, 3,  0, 1, 23, 47, 23, 18],
#     "Jkt_2":        [1,  2, 3,  2, 0, 15, 30, 12, 14],
#     "Bali":         [1,  1, 1,  4, 3,  0, 32, 14, 11]
# }

data = {
    "West Java":    [0, 11, 5, 10, 7, 48],
    "Central Java": [2,  0, 2, 10, 6, 29],
    "East Java":    [4,  7, 0,  7, 5, 30],
    "Grtr. Jkt_1":        [0,  4, 3,  0, 1, 23],
    "Grtr. Jkt_2":        [1,  2, 3,  2, 0, 15],
    "Bali":         [1,  1, 1,  4, 3,  0]
}
# Convert the data into a DataFrame
# df = pd.DataFrame(data, index=["West Java", "Central Java", "East Java", "Jkt_1", "Jkt_2", "Bali", "GPT-3.5", "GPT-4", "GPT-4-t"])
df = pd.DataFrame(data, index=["West Java", "Central Java", "East Java", "Grtr. Jkt_1", "Grtr. Jkt_2", "Bali"])

# Plotting the heatmap
plt.figure(figsize=(13, 7))
cmap = sns.color_palette("crest", as_cmap=True)
ax = sns.heatmap(df, annot=True, fmt="d", cmap=cmap, annot_kws={"size": 28}, vmin=0, vmax=48)
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=26)

# Titles and labels
# plt.title('Answer Conflict in Indonesian Data', fontsize=18)
plt.xlabel('Question Creator', fontsize=28)
plt.ylabel('Answerer', fontsize=28)

# Adjust the font size of the tick labels
plt.xticks(fontsize=26)
plt.yticks(fontsize=26)
plt.xticks(rotation=25)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('figures/anno_conflict_ind.pdf', format='pdf', bbox_inches="tight")
plt.show()
```
