```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "West Java": [0, 11, 5, 10, 7, 48, 90, 35],
    "Central Java": [2, 0, 2, 10, 6, 29, 57, 28],
    "East Java": [4, 7, 0, 7, 5, 30, 48, 22],
    "Betawi1": [0, 4, 3, 0, 1, 23, 50, 23],
    "Betawi2": [1, 2, 3, 2, 0, 15, 28, 13],
    "Bali": [1, 1, 1, 4, 3, 0, 32, 13]
}

# Convert the data into a DataFrame
df = pd.DataFrame(data, index=["West Java", " Central Java", " East Java", "Betawi1", "Betawi2", "Bali", "GPT-3.5", "GPT-4"])

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, fmt="d", cmap="coolwarm", annot_kws={"size": 17})

# Titles and labels
plt.title('Answer Conflict in Indonesian Data', fontsize=18)
plt.xlabel('Question Generator', fontsize=17)
plt.ylabel('Answerer', fontsize=17)

# Adjust the font size of the tick labels
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.show()
```
